```python
import re

UN = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
DB={'UserName':[],
   'Password':[]}
#index=0

def checkE(email):#checking mail and register
    if(re.fullmatch(UN, email)):
        Un=email
        dump=input('Enter Password ')
        checkP(dump)
        if checkP(dump)==1:
            print('registered')
            Up=dump
            DB['UserName'].append(Un)
            DB['Password'].append(Up)
    else:
        print("Invalid Email, Try again ")
        NE=input()
        checkE(NE)

def checkP(password):#checking valid password
    l, u, p, d = 0, 0, 0, 0
    capitalalphabets="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    smallalphabets="abcdefghijklmnopqrstuvwxyz"
    specialchar="$@_~!#%^&*"
    digits="0123456789"
    if (len(password)>5 and len(password)<16):
        for i in password:
            if (i in smallalphabets):
                l+=1           
            if (i in capitalalphabets):
                u+=1           
            if (i in digits):
                d+=1           
            if(i in specialchar):
                p+=1       
        if (l>=1 and u>=1 and p>=1 and d>=1 and l+p+u+d==len(password)):
            return 1
        else:
            print("Invalid Password, Try again ")
            NP=input()
            checkP(NP)
    if(len(password)<5 or len(password)>16):
        print("Invalid Password, Try again ")
        NPP=input()
        checkP(NPP)

def login(emid):#checking login credentials
    if emid in DB['UserName']:
        index=DB['UserName'].index(emid)
        print('Enter password ')
        PWD=input()
        def PWcheck(PW):
            if PW==DB['Password'][index]:
                print('login successfull ')
            else:
                print('wrong password')
                print('press 1 to try again\npress 2 if forgot password')
                choice1=int(input())
                if choice1==1:
                    print('Enter Password')
                    PW2=input()
                    PWcheck(PW2)
                if choice1==2:
                    print('Confirm Email ID ')
                    ID2=input()
                    index2=DB['UserName'].index(ID2)
                    if ID2 in DB['UserName']:
                        print('Your Password is\n')
                        print(DB['Password'][index2])
        PWcheck(PWD)
    else:
        print('User not found enter valid ID and register')
        Nuser=input()
        checkE(Nuser)



#driver code       
print('press 1 to register\npress 2 to login')
choice=int(input())
if choice==1:
    print('Enter Valid ID ')
    ID=input()
    checkE(ID)
if choice==2:
    print('Enter Your ID ')
    EID=input()
    login(EID)
```
